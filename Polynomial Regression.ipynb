### Polynomial Regression

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data={
    'Study_Hours': [1,2,3,4,5],
    'Marks': [10,25,55,90,130]
      }
df=pd.DataFrame(data)
df
```

```python
X=df[['Study_Hours']]
y=df['Marks']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_poly,y_train)
```

```python
y_train_pred = model.predict(X_train_poly)
y_test_pred = model.predict(X_test_poly)
```

```python
print("Training MSE:", mean_squared_error(y_train,y_train_pred))
print("Testing MSE:", mean_squared_error(y_test,y_test_pred))
```

```python
plt.scatter(X,y,color="blue",label="actual data")
xline=np.linspace(min(X["Study_Hours"]),max(X["Study_Hours"]),100).reshape(-1,1)
yline=model.predict(poly.transform(xline))
plt.plot(xline,yline,color='red',label="polynomial fit")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("polynomail regression")
plt.legend()
plt.show()
```
