## Covariance
covariance tells you whether two variables move together.

positive covariance --> X variable increases y also increases
Negative covariance --> X variable increases y also decreases
Zero covariance --> no relation b/w the variables
NOTE:: covariance shows only direction not strength

```python
import pandas as pd
df=pd.read_csv('study_marks.csv')
df
```

```python
df.var()
```

```python
df.corr()
```

```python
df.cov()
```

```python
data= {
    "temp": [30,32,34,35,36],   # negative correlation
    "humidity":[40,39,34,28,27]
}
df=pd.DataFrame(data)
df
```

```python
df.cov()
```

```python
df.var()
```

```python
df.corr()
```
