### What is ROC Curve?
 ROC(Receiver Operating Characteristic) curve is a graph that shows how well a binary classification model can seperate positive and negative classes at different threshold values.
     It plots:
    >> X-axis : False Positive Rate (FPR)
    >> Y-axis : true Positive Rate(TPR)
### What is AUC Curve?
  AUC(Area Under the ROC Curve) is a single number that summarizes the ROC curve.
  AUC = 1.0 --> Perfect model
  AUC = 0.5 --> Random guessing
  Higher AUC = Better model
  Example: Disease--> +ve(1)
           NO disease--> -ve(0)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
y_true = np.array([1,1,0,0,1]) #actual 
y_scores = np.array([0.95,0.85,0.70,0.40,0.30]) # predicted
```

```python
fpr,tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)
print("AUC:", roc_auc)
```

```python
plt.figure()
plt.plot(fpr, tpr, label=f'ROC curve(AUC = {roc_auc:.2f}')
plt.plot([0,1],[0,1], linestyle = '--') # random line
plt.xlabel("false Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC curve")


```
