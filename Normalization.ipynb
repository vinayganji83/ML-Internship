## Normalization
### -Normalization is a technique used to scale the values of features into small range,usually 0 to 1.
### -it makes all features equal in scale so the model does not get confused by big and small numbers.

## why do we use normalisation
### Features have different ranges A is equal to 20 - 60 
### -Big numbers dominate small numbers.
### - Models work better when all values are in the same range.
### - Training become faster and more accurate.

### Normalisation is mainly needed in KNN, K means clustering, Neural networks, svm, Logistic regression.


```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
s = pd.Series([1000,2000,3000,4000,5000])
```

```python
s
```

```python
s.min()
```

```python
s.max()
```

```python
s_norm = (s-s.min())/(s.max()-s.min())
```

```python
print(s_norm)
```

```python
s_norm = MinMaxScaler().fit_transform(s.values.reshape(-1,1)) #using Library
print(s_norm)
```
