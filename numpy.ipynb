# copies and viewing

```python
a=np.arange(10)
b=a[::2]
print(a)
print(b)
```

```python
print(np.shares_memory(a,b))
```

```python
b[0]=10 #viewing
print(a)
print(b)
```

```python
a=np.arange(10)
b=a[::2].copy() #force a copy
print(a)
print(b)
```

```python
print(np.shares_memory(a,b))
```

# Random number genarator

```python
mask=(a%2==0)
```

```python
mask_a = a[mask]
print(mask_a)
```
